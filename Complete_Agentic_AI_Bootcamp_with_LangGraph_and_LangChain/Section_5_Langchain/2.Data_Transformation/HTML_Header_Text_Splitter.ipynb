```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_template="""
<!DOCTYPE html>
<html>
<head>
    <title>Data Engineering Overview</title>
</head>
<body>

<h1>Introduction to Data Engineering</h1>
<p>Data engineering is the practice of designing and building systems for collecting, storing, and analyzing data.</p>

<h2>What is Data Ingestion?</h2>
<p>Data ingestion is the process of collecting raw data and moving it into a storage or processing system.</p>
<p>It can include batch processing, real-time streaming, and hybrid approaches.</p>

<h3>Batch Ingestion</h3>
<p>Batch ingestion collects data over a period of time and ingests it all at once.</p>

<h3>Real-Time Ingestion</h3>
<p>Real-time ingestion processes data as soon as it arrives, often using tools like Kafka or Kinesis.</p>

<h2>Data Transformation</h2>
<p>Transformation includes cleaning, normalizing, joining, aggregating, and enriching data.</p>

<h3>ETL</h3>
<p>ETL stands for Extract, Transform, Load. It is used in traditional data warehousing.</p>

<h3>ELT</h3>
<p>ELT stands for Extract, Load, Transform and is used in modern cloud data platforms.</p>

<h2>Data Storage</h2>
<p>Data can be stored in data lakes, data warehouses, or lakehouses.</p>

<h3>Data Lakes</h3>
<p>Data lakes can store raw and unstructured data, usually on cheap storage.</p>

<h3>Data Warehouses</h3>
<p>Data warehouses store structured data optimized for analytics.</p>

<h1>Conclusion</h1>
<p>Data engineering is essential for modern analytics and machine learning.</p>

</body>
</html>
"""
```

```python
headers_to_split=[
    ("h1","Header 1"),
    ("h2","Header 2"),
    ("h3","Header 3")
]

splitter=HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split)
split_docs=splitter.split_text(html_template)
print(split_docs[0])
```

```python
url="https://plato.stanford.edu/support/"

headers_to_split_on=[
    ("h1","Header 1"),
    ("h2","Header 2"),
    ("h3","Header 3")
]

splitter=HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
split_docs=splitter.split_text_from_url(url)
print(split_docs[0].page_content)
```
