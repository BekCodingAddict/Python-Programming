```python
import os 
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import OpenAIEmbeddings
emmbeddings = OpenAIEmbeddings(model="text-embedding-3-large")
emmbeddings
```

```python
text="This is OpenAI Embedding Example"
query_result = emmbeddings.embed_query(text)
query_result
```

```python
print(query_result[0])
len(query_result)
```

```python
emmbeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
emmbeddings_1024
```

```python
text="This is OpenAI Embedding Example"
query_result = emmbeddings_1024.embed_query(text)
print(query_result)
len(query_result)
```

```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader(file_path="What_is_DataIngestion.txt")
text_docs=loader.load()
for doc in text_docs:
    print(doc.page_content)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_docs=text_splitter.split_documents(text_docs)
for doc in final_docs:
    print(doc.page_content)
```

```python
## Vecotr Embedding And Vector StroreDB Chroma
from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_docs,emmbeddings_1024)
db
```

```python
query="Key Aspects"
retrived_result=db.similarity_search(query)
for doc in retrived_result:
    print("Source:", doc.metadata.get("source"))
    print("Content:", doc.page_content)
    print("-" * 80)
```
