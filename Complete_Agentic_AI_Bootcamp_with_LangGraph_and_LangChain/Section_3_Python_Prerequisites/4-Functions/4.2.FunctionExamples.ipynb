```python
def convert_to_temerature(celsius):
    """ 
    Convert Celsius to Fahrenheit
    """
    return (celsius * 9/5) + 32

print(convert_to_temerature(20))
```

```python
## Passowrs Strength Checker
def check_password(password):
    """
    Check password strength
    """
    if len(password) < 8:
        return "Weak"
    if len(password) >= 8 and any(char.isupper() for char in password):
        return "Medium"
    if len(password) >= 8 and any(char.islower() for char in password):
        return "Medium"
    if len(password) >= 8 and any(char.isdigit() for char in password):
        return "Medium"
    if len(password) >= 8 and any(char in "!@#$%^&*()_+" for char in password):
        return "Strong"
    return "Weak"

print(check_password("Password123"))
print(check_password("password123"))
print(check_password("Password123!"))
print(check_password("Pass"))
```

```python
## Canculate the total  cost of items in a shopping cart
def calculate_total_cost(cart):
    """
    Calculate the total cost of items in a shopping cart
    """
    total = 0
    for item in cart:
        total += item["price"] * item["quantity"]
    return total


cart = [
    {"item": "apple", "price": 1.5, "quantity": 10},
    {"item": "banana", "price": 2.5, "quantity": 5},
    {"item": "orange", "price": 3.5, "quantity": 3}
]

print(calculate_total_cost(cart))
```

```python
def is_palindrome(word):
    """
    Check if a word is a palindrome
    """
    return word == word[::-1]

print(is_palindrome("racecar"))
print(is_palindrome("hello"))
```

```python
## Recursive Factorial
def factorial(n):
    """
    Calculate the factorial of a number
    """
    if n == 0:
        return 1
    return n * factorial(n-1)

print(factorial(7))
```

```python
def count_word_frequency(text):
    """
    Count the frequency of each word in a text
    """
    words = text.split()
    frequency = {}
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1
    return frequency

text = "Hello world, hello everyone. Welcome to the world of Python."
print(count_word_frequency(text))
```

```python
import os
print(os.getcwd())

def count_word_frequency_from_file  (file_path):
    """
    Count the frequency of each word in a file
    """
    frequency = {}
    with open(file_path, "r") as file:
        text = []
        for line in file:
            text += line.split()
            for word in text:
                word = word.lower().strip(".,!?")
                frequency[word] = frequency.get(word, 0) + 1
    return frequency

file_path = os.path.join(os.getcwd(), "sample.txt")
print(file_path)
print(count_word_frequency_from_file(file_path))
```

```python
import re
def is_valid_email(email):
    """
    Check if an email is valid
    """
    return re.match(r"[^@]+@[^@]+\.[^@]+", email) is not None

print(is_valid_email("test@example.com"))
print(is_valid_email("test@example"))
print(is_valid_email("test@example.com.au"))
```
