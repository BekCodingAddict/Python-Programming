```python
my_list = [1, 2, 3, 4, 5]
for i  in my_list:
    print(i)
```

```python
iterator=iter(my_list)
while True:
    try:
        print(next(iterator))
    except StopIteration:
        break
```

```python
string="Hello"
iterator=iter(string)
while True:
    try:
        print(next(iterator))
    except StopIteration:
        print("No more elements")
        break


```
