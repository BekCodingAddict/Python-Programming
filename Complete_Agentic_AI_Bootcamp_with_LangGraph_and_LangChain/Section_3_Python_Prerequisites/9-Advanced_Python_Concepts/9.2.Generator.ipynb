```python
def square(n) -> int:
    for i in range(n):
        yield i**2

squares=square(3)

for i in squares:
    print(i)

```

```python
while True:
    try:
        print(next(squares))
    except StopIteration:
        print("No more elements")
        break
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3

gen=my_generator()

for i in gen:
    print(i)
```

```python
# Practical Example: Reading Large Files
def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line

file_path = 'large_file.txt'

for line in read_large_file(file_path):
    print(line)

```
