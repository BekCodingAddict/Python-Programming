```python
## Read Whole File
with open("data.txt","r") as file:
    content=file.read()
    print(content)
```

```python
## Read Line by Line
with open("data.txt","r") as file:
    print(type(file))
    for line in file:
        print(line)

```

```python
## Write to a File
with open("data.txt","w") as file:
    file.write("Hello, World!\n")

## Append to a File
with open("data.txt","a") as file:
    file.write("Python is a high-level, general-purpose programming language.")

```

```python
## Read List of line to a file
lines=["Hello, World!\n","Python is a high-level, general-purpose programming language.\n","Python is a high-level, general-purpose programming language.\n"]
with open("data.txt","w") as file:
    file.writelines(lines)

```

```python
## Binary File
data=b'\x00\x01\x02\x03\x04\x05'
with open("example.bin","wb") as file:
    file.write(data)
```

```python
##Read Binary File
with open("example.bin","rb") as file:
    data=file.read()
    print(data)
```

```python
## Read the content of the file and write to a destination file
with open("data.txt","r") as source:
    content=source.read()

with open("destination.txt","w") as destination:
    destination.write(content)

```

```python
## Writing and Reading  a file

with open("data.txt","w+") as file:
    file.write("Hello, World!\n")
    file.write("Python is a high-level, general-purpose programming language.\n")
    ## Move the cursor to the beginning of the file
    file.seek(0)

    content=file.read()
    print(content)
```
