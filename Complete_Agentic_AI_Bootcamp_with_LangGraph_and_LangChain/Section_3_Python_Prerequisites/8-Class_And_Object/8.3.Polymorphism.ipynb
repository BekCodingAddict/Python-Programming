# Polymorphism
### Method Overriding 

```python
class Smartphone:
    def __init__(self,brand,model,price):
        self.brand=brand
        self.model=model
        self.price=price

    def get_info(self):
        return f"The mobile phone brand is {self.brand}, model is {self.model} and price is {self.price}"

class Android(Smartphone):
    def get_info(self):
        return f"The android phone brand is {self.brand}, model is {self.model} and price is {self.price}"

class Iphone(Smartphone):
    def get_info(self):
        return f"The iphone phone brand is {self.brand}, model is {self.model} and price is {self.price}"

Samsung=Android("Samsung","Galaxy S20",500)
print(Samsung.get_info())
Iphone=Iphone("Apple","iPhone 12",700)
print(Iphone.get_info())
```

### Polymorphism with Abstract Base Class
Abstract Base Class is a class that is designed to be a base class for other classes.
- It is a class that is not intended to be instantiated directly.
- It is a class that is intended to be a base class for other classes.
   

```python
from abc import ABC, abstractmethod

## define a class that is a base class for other classes
class Smartphone(ABC):
    def __init__(self,brand,model,price):
        self.brand=brand
        self.model=model
        self.price=price
    @abstractmethod
    def get_info(self):
        pass

class Android(Smartphone):
    def get_info(self):
        return f"The android phone brand is {self.brand}, model is {self.model} and price is {self.price}"

class Iphone(Smartphone):
    def get_info(self):
        return f"The iphone phone brand is {self.brand}, model is {self.model} and price is {self.price}"

iphone=Iphone("Apple","iPhone 12",700)
print(iphone.get_info())
samsung=Android("Samsung","Galaxy S20",500)
print(samsung.get_info())
```
