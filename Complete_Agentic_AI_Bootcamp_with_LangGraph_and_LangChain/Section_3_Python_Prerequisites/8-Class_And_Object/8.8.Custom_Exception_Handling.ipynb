```python
class Error(Exception):
    pass

class dobException(Exception):
    pass
```

```python
year = int(input("Enter the year of birth: "))
age = 2025 - int(year)

try:   
    if age <=30  and age >= 18:
        print("You are eligible to apply for the job")
    else:
        raise dobException("You are not eligible to apply for the job")
    
except dobException:
    print("You are not eligible to apply for the job")
except Error:
    print("Invalid date of birth")
```
