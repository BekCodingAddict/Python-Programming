```python
class Car:
    def __init__(self,name,model,year):
        self.name=name
        self.model=model
        self.year=year

    def get_info(self):
        return f"The car name is {self.name}, model is {self.model} and year is {self.year}"
       
class ElectricCar(Car):
    def __init__(self,name,model,year,battery_size):
        super().__init__(name,model,year)
        self.battery_size=battery_size

    def get_info(self):
        return f"The car name is {self.name}, model is {self.model} and year is {self.year} and battery size is {self.battery_size}"
    
    
BMW=Car("BMW","320",2021)
print(BMW.get_info())

Tesla=ElectricCar("Tesla","Model S",2020,100)
print(Tesla.get_info())
```

```python
## Multiple Inheritance
class Animal:
    def __init__(self,name):
        self.name=name

    def speak(self):
        print("Subclass must implement this method")

class Pet:
    def __init__(self,age):
        self.age=age

    
class Cat(Animal,Pet):
    def __init__(self,name,age,color):
        Animal.__init__(self,name)
        Pet.__init__(self,age)
        self.color=color

    def speak(self):
        print(f"{self.name} is meowing")

Cat=Cat("Whiskers",2,"White")
Cat.speak()
```
