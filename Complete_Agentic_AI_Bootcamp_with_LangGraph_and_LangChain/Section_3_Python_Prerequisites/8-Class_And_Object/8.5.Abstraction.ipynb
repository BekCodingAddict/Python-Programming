```python
from abc import ABC, abstractmethod

## Abstraction is the process of hiding the implementation details of a class and showing only the necessary information to the user.
class Laptop(ABC):
    @abstractmethod
    def details(self):
        pass

class Windows(Laptop):
    def __init__(self, name, ram, cpu, storage):
        self.__name = name
        self.__ram = ram
        self.__cpu = cpu
        self.__storage = storage

    def details(self):
        print(f"Name: {self.__name},\n RAM: {self.__ram},\n CPU: {self.__cpu},\n Storage: {self.__storage}")

class Macbook(Laptop):
    def __init__(self, name, ram, cpu, storage):
        self.__name = name
        self.__ram = ram
        self.__cpu = cpu
        self.__storage = storage

    def details(self):
        print(f"Name: {self.__name},\n RAM: {self.__ram},\n CPU: {self.__cpu},\n Storage: {self.__storage}")

Laptop1 = Windows("Windows", "8GB", "Intel i5", "1TB")
Laptop1.details()
MacLaptop = Macbook("Macbook", "16GB", "Apple M1", "512GB")
MacLaptop.details()
```
