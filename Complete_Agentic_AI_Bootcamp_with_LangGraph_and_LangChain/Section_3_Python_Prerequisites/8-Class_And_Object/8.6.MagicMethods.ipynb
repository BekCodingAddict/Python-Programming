## Magic Methods are the methods that are used to change the behavior of the class.

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __str__(self):
        return f"Employee name is {self.name} and salary is {self.salary}"

    def __repr__(self):
        return f"Employee(name={self.name}, salary={self.salary})"

    def __add__(self, other):
        return self.salary + other.salary

    def __len__(self):
        return len(self.name)

    def __del__(self):
        print("Object is deleted")

    def __call__(self):
        print("Object is called")

    def __getitem__(self, index):
        return self.name[index]

    def __setitem__(self, index, value):
        self.name[index] = value

    def __contains__(self, item):
        return item in self.name

    def __iter__(self):
        return iter(self.name)

    def __next__(self):
        return next(self.name)
    

    def __getattr__(self, attr):
        return f"No attribute found: {attr}"

    def __setattr__(self, attr, value):
        self.__dict__[attr] = value

    def __delattr__(self, attr):
        del self.__dict__[attr]

    def __getattribute__(self, attr):
        return f"Getting attribute: {attr}"

    def __setattribute__(self, attr, value):
        self.__dict__[attr] = value
  
    
    def __get__(self, obj, objtype):
        return f"Getting attribute: {obj}"
```
