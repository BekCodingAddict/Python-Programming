```python
## Class Creation
class Car:
    def __init__(self,name,model,year):
        self.name=name
        self.model=model
        self.year=year
        
    def get_info(self):
        return f"The car name is {self.name}, model is {self.model} and year is {self.year}"

Audi=Car("Audi","A4",2020)
print(Audi.get_info())
print(type[str, str, int](Audi))
```

```python
class programming_language:
    def __init__(self,name,creator,year):
        self.name=name
        self.creator=creator
        self.year=year

        self._programming_languages=[{
            "name":"Python",
            "creator":"Guido van Rossum",
            "year":1991,
            "is_dynamic":True,
            "is_static":False,
            "is_interpreted":True
        },
        {
            "name":"Java",
            "creator":"James Gosling",
            "year":1995,
            "is_dynamic":False,
            "is_static":True,
            "is_interpreted":False
        },
        {
            "name":"C++",
            "creator":"Bjarne Stroustrup",
            "year":1985,
            "is_dynamic":False,
            "is_static":True,
            "is_interpreted":False
        },
        ]
        
    def get_info(self):
        return f"The programming language name is {self.name}, creator is {self.creator} and year is {self.year}"

    def is_dynamic(self):
        for language in self._programming_languages:
            if language["name"]==self.name:
                return language["is_dynamic"]
        return False
     
    def is_static(self):
        for language in self._programming_languages:
            if language["name"]==self.name:
                return language["is_static"]
        return False
        
    def is_interpreted(self):
        for language in self._programming_languages:
            if language["name"]==self.name:
                return language["is_interpreted"]
        return False
    

Python=programming_language("Python","Guido van Rossum",1991)
print(Python.is_dynamic())
print(Python.is_static())
print(Python.is_interpreted())
    
    
    
    
```
