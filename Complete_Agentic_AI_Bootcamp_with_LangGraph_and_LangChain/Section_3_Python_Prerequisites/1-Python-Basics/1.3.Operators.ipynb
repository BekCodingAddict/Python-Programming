```python
## Arithmetic Operators
a=10
b=20

sum=a+b ## Addition
sub_result=b-a ## Subtraction
mul_result=a*b ## Multiplication
div_result=b/a ## Division
mod_result=b%a ## Modulus
floor_div_result=b//a ## Floor Division
modulus_result=b%a ## Exponentiation
exponent_result=b**a ## Exponentiation


print(sum)
print(sub_result)
print(mul_result)
print(div_result)
print(mod_result)
print(floor_div_result)
print(modulus_result)
print(exponent_result)  




```

```python
## Comparison Operators
## Equal to
a=10
b=20
print(a==b)

str1="Hello"
str2="Hello"
print(str1==str2)
```

```python
## Not Equal to
str1!=str2
```

```python
## Greater than
a=10
b=20
print(a>b)
## Less than
print(a<b)

```

```python
## greater than or equal to

a=10
b=20
print(a>=b)

## less than or equal to
print(a<=b)

```

```python
## Logical Operators
## AND, OR, NOT
x=True
y=False

print(x and y)
print(x or y)
print(not x)

```
