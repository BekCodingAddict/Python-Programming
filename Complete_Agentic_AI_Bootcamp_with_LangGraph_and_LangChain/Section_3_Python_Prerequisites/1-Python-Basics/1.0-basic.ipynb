Syntax refers to the set of rules that defines the combinations of symbols that are considered to be correctly structured programs in a language. In simpler terms,syntax is about the correct argument of words and symbols in a code. 

Semantics refers to  the meaning or the interpretation of the symbols,characters and commands in a language. it is about what the code is supposed to when it runs. 

```python
## Basic Syntax Rules in Python
## Case Sensetivity -Python is case sensitive

name="John"
Name="Jane"
print(name)
print(Name)
```

```python
## Line Continuation -Python allows line continuation using the backslash character

total=100+200+300+400+500\
        +600+700+800+900+1000
print(total)
```

```python
## Multiple Statements on a Single Line -Python allows multiple statements in a single line separated by semicolons

x=10;y=20;z=30
print(x,y,z)
```
