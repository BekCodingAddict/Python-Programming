```python
## Declaring and Assigning Variables

age=32
name="John"
is_student=True
height=5.9

## printing variables
print("age:",age)
print("name:",name)
print("is_student:",is_student)
print("height:",height)
```

## Naming Conventions
-  Variable names should be in lowercase
-  Variable names should be descriptive
-  Variable names should not be keywords
-  Variable names should not be special characters
-  Variable names should not be numbers
-  Variable names should not be too long
-  Variable names should not be too short
-  Variable names should not be too complex
-  Variable names should not be too generic
-  Variable names should not be too specific
-  Variable names should not be too vague
-  Variable names should not be too similar to each other

```python
## Understanding Variables types
## Python is a dynamically typed language

age=32 #int
name="John" #str
is_student=True #bool
height=5.9 #float

## Printing variables
print(type(age) )
print(type(name))
print(type(is_student))
print(type(height))

## Type Casting

```

```python
## Type Conversion

age=32

age_str=str(age)
print(type(age_str))

## Converting float to int
height=5.9
height_int=int(height)
print(type(height_int))
```

```python
## Dynamic Typing
## Python allows the type of variable to be changed dynamically as executed
var=10
print(var,type(var))
var=10.5
print(var,type(var))
var="Hello, World!"
print(var,type(var))

```
