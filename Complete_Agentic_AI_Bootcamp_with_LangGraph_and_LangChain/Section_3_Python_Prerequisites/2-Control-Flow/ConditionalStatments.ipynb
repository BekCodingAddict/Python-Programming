```python
## if statement

age=18

if age>=18:
    print("You are an adult")
else:
    print("You are not an adult")
    
    
```

```python
## Nested Conditional Statements

score=85

if score>=90:
    print("Grade: A")
    if score>=95:
        print("Excellent")
elif score>=80:
    print("Grade: B")
    if score>=85:
        print("Good")
else:
    print("Grade: C")


```

```python
num1=float(input("Enter the first number: "))
operator=input("Enter the operator (+, -, *, /):")
num2=float(input("Enter the second number: "))

if operator=="+":
    result=num1+num2
elif operator=="-":
    result=num1-num2
elif operator=="*":
    result=num1*num2
elif operator=="/":
    if num2!=0:
        result=num1/num2
    else:
        result="Error: Division by zero"
else:
    result="Error: Invalid operator"
    
print(f"The result is: {result}")
```
