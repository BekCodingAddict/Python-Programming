```python
## for loop

for i in range(5):
    print(i)
print("Done")
```

```python
## String 

str = "Hello World"
for i in str:
    print(i)
print("Done")
```

```python
## while loop

i = 0
while i < 5:
    print(i)
    i += 1
print("Done")


```

```python
 ## Loop Control Statements
# break

for i in range(10):
    if i == 5:
        break
    print(i)
print("Done")
```

```python
## continue

for i in range(10):
    if i%2 == 0:
        continue
    print(i)
print("Done")

```

```python
## pass

for i in range(10):
    if i%2 == 0:
        pass
    else:
        print(i)
print("Done")
```

```python
## Nested Loops

for i in range(3):
    for j in range(3):
        print(i, j)
print("Done")
```

```python
## for loop with else

for i in range(5):
    if i == 3:
        pass
    else:
        print(i)
else:
    print("Done")
```
