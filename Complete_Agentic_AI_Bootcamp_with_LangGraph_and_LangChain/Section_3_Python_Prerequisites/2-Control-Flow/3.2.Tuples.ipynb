```python
## create a tuple
from typing import Literal


my_tuple=("apple","banana","cherry")
print(my_tuple)
print(type[Literal['apple', 'banana', 'cherry'], ...](my_tuple))
print(type(Literal))
```

```python
from typing import Any


my_tuple=list[Any]()
tpl=tuple[Any, ...]()
print(type[Any, ...](tpl))
```

```python
numbers=tuple[int,...]([1,2,3])
print(numbers)
print(type[int, ...](numbers))
```

```python
mixed_tuple=tuple[int,float,str]([1,2.0,"three"])
print(mixed_tuple)
print(type[int,float,str](mixed_tuple))
print(mixed_tuple[0])
```

```python
## Accessing elements
print(mixed_tuple[1])
print(mixed_tuple[-2])
```

```python
numbers=tuple[int,...]([1,2,3,4,5])
print(numbers)
print(type[int,...](numbers))
numbers=tuple[int,...]([1,2,3,4,5])
print(numbers)
print(type[int,...](numbers))
```

```python
print(numbers[0:4])
print(numbers[::-1])
```

```python
## Tuple operations
mixed_tuple=tuple[int,float,str]([1,2.0,"three"])
numbers=tuple[int,...]([1,2,3,4,5])
concatenated=numbers+mixed_tuple
print(concatenated)
print(type[int,...](concatenated))
print(concatenated*3)



```

```python
#Immutability
print(mixed_tuple.count(1))
print(concatenated.count(1))
print(mixed_tuple.index(2))
```

```python
##Packing and Unpacking
packed_tuple=1,"Hello",3.14
print(packed_tuple)

a,b,c=packed_tuple
print(a,b,c) 

numbers=tuple[int,...]([1,2,3,4,5])
first,*middle,last=numbers
print(first)
print(middle)
print(last)





```
