```python
num = 10
def isOdd(num):
    return "Odd" if num % 2 != 0 else "Even"

print(isOdd(num))

```

```python
## Function with multiple parameters
def add_num(num1: int, num2: int):
    return num1 + num2

print(add_num(10, 20))
```

```python
## Default parameters
def add_num(num1: int, num2: int = 10):
    return num1 + num2

print(add_num(10))
```

```python
## Variable length parameters
## positional arguments and keyword arguments
## positional arguments are passed to the function as a tuple
## keyword arguments are passed to the function as a dictionary
def add_num(*args):
    return sum(args)

print(add_num(10, 20, 30, 40, 50))
```

```python
## Keyword arguments
def show_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")

print(show_details(name="John", age=30, city="New York"))
```

```python
def print_details(*args, **kwargs):
    for arg in args:
        print(arg)

    for key, value in kwargs.items():
        print(f"{key} : {value}")

print_details("John", "Doe", name="John", age=30, city="New York")
```

```python
## Return statement

def multiply_num(*args: int):
    result = 1
    for num in args:
        result *= num
    return result

print(multiply_num(1, 2, 3, 4, 5))
```
