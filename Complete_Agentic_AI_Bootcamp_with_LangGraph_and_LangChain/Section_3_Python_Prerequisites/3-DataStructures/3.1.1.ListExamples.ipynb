```python
to_do_list = ["Buy groceries", "Finish the course", "Buy a new phone"]

## Adding to the list
to_do_list.append("Buy a new laptop")
to_do_list.append("Do the laundry")

## Removing from the list
to_do_list.remove("Buy groceries")

## checking if an item is in the list
if "Buy groceries" in to_do_list:
    print("You need to buy groceries")
   
print("To do list remaining:")
for item in to_do_list:
    print(item)
   
```

```python
## Organizing students by grade
students = [
    {"name": "John", "grade": "A"},
    {"name": "Jane", "grade": "B"},
    {"name": "Jim", "grade": "C"},
]

## Sorting students by grade
students.sort(key=lambda x: x["grade"])

print("Students sorted by grade:")
for student in students:
    print(student["name"], " - ", student["grade"])



    
```
