```python
df.rename(columns={"purchase_amount":"purchase_amount_new"},inplace=True)
```

```python
df.head()
```

```python
## Change the data type of a column
df["Value_new"]=df["age"].astype(float)
df.head()
df.dtypes

## Drop a column
df.drop(columns=["Value_new"],inplace=True)
df.head()
```

```python
df['New Value']=df['age'].apply(lambda x: x*2)
df.head()

```

```python
## Data Aggregation and Grouping
grupped_mean=df.groupby("city")["purchase_amount_new"].mean()
grupped_mean.to_frame()
```

```python
grupped_sum=df.groupby(["city","name"])["purchase_amount_new"].sum()
grupped_sum.to_frame()
```

```python
## Merging and Joining DataFrames
df1=pd.DataFrame({"id":[1,2,3,4,5],"name":["Alice","Bob","Charlie","David","Eva"]})
df2=pd.DataFrame({"id":[1,2,3,4,5],"age":[25,30,35,40,45]})
df1.head()
```

```python
df2.head()
```

```python
## Merge the two dataframes on the "key" column
pd.merge(df1,df2,on="id",how="inner")
```

```python
pd.merge(df1,df2,on="id",how="left")
```

```python
pd.merge(df1,df2,on="id",how="right")
```
