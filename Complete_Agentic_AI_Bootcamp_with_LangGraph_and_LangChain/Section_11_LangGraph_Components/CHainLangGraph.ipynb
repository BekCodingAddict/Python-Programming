```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
import os

load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

## How to pass sequense of massage to AI LLM model 

```python
from langchain_core.messages import HumanMessage,AIMessage
from pprint import pprint

messages=[AIMessage(content="Hello, how are you?",name="LLMModel")]
messages.append(HumanMessage(content="I'm good, thank you!",name="User"))
messages.append(AIMessage(content="I'm also good, thank you!",name="LLMModel"))
messages.append(HumanMessage(content="I wanna learn about the topic of Agenitc AI",name="User"))

for message in messages:
    message.pretty_print()
```

```python
from langchain_groq import ChatGroq
llm=ChatGroq(model_name="llama-3.1-8b-instant",api_key=os.getenv("GROQ_API_KEY"))
result=llm.invoke(messages)
result.pretty_print()
```
